# Employment in the AI Era - Analysis of Automation Risk and Job Growth Trends

### Introduction

Artificial Intelligence (AI) is becoming an integral part of modern workplaces, transforming how businesses operate and how employees perform their jobs. While AI increases efficiency and productivity, it also raises concerns about job automation, workforce displacement, and changing skill requirements. At the same time, AI is creating new opportunities and driving demand for emerging roles across various industries.

This project explores the impact of AI on future jobs using a comprehensive dataset containing information on job roles, industries, salaries, automation levels, hiring trends, and employee-related factors. Through data cleaning, exploratory data analysis (EDA), and visualization techniques, the project aims to identify patterns, trends, and relationships that reveal how AI is shaping the future workforce. The insights derived from this analysis can help organizations, employees, and policymakers better understand and prepare for the evolving job market.

### Problem Definition

Artificial Intelligence is rapidly transforming industries by automating tasks, improving operational efficiency, and changing workforce requirements. While some jobs are becoming more vulnerable to automation, others are experiencing increased demand due to technological advancements. Organizations and employees face uncertainty regarding which skills will remain valuable, which industries will continue to grow, and how AI adoption will influence employment opportunities, salaries, and job satisfaction.

The problem addressed in this project is to analyze the impact of Artificial Intelligence on future jobs by examining factors such as AI replacement risk, automation levels, future demand scores, hiring trends, salaries, and employee performance. Through exploratory data analysis and visualization, this project aims to identify patterns and insights that can help understand how AI is reshaping the labor market and what factors contribute to career sustainability in an AI-driven economy.

### Outliers

Outliers were identified using boxplots and the Interquartile Range (IQR) method. Detecting outliers helps determine whether extreme values may influence the analysis.

```python
# Detect Outliers

numerical_cols = [
    'Years_Experience',
    'AI_Replacement_Risk',
    'Future_Demand_Score',
    'Average_Salary_USD',
    'Job_Growth_2030',
    'Work_Hours_Per_Week',
    'Performance_Score',
    'Job_Satisfaction'
]
```

```python
# Visualize Outliers Using Boxplots

plt.figure(figsize=(15,10))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Count Outliers Using IQR Method

for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    print(f"{col}: {len(outliers)} outliers")
```

```python
# Create an Outlier Summary Table

outlier_summary = []

for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = len(
        df[
            (df[col] < lower) |
            (df[col] > upper)
        ]
    )

    outlier_summary.append([col, count])

outlier_df = pd.DataFrame(
    outlier_summary,
    columns=['Column','Outlier Count']
)

outlier_df
```

##### Outlier analysis was performed using the Interquartile Range (IQR) method. Although some outliers were detected in salary, experience, and performance-related variables, they were retained because they represent realistic workforce variations and provide meaningful business insights.
