## DATA CLEANING & PREPROCESSING

```python
# Check Missing Values

df.isnull().sum()
```

```python
# Visualize Missing Values

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Missing Value Percentage

missing_percent = (df.isnull().sum()/len(df))*100

missing_df = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing Percentage': missing_percent
})

missing_df[missing_df['Missing Count']>0]
```

#### Handling Missing Values

```python
# Years Experience

df['Years_Experience'] = df['Years_Experience'].fillna(df['Years_Experience'].median())
```

```python
df['Years_Experience'].isnull().sum()
```

```python
# AI Replacement Risk

df['AI_Replacement_Risk'] = df['AI_Replacement_Risk'].fillna(df['AI_Replacement_Risk'].median())
```

```python
df['AI_Replacement_Risk'].isnull().sum()
```

```python
# Job Growth 2030

df['Job_Growth_2030'] = df['Job_Growth_2030'].fillna(
    df['Job_Growth_2030'].median()
)
```

```python
df['Job_Growth_2030'].isnull().sum()
```

```python
# Work Hours Per Week

df['Work_Hours_Per_Week'] = df['Work_Hours_Per_Week'].fillna(
    df['Work_Hours_Per_Week'].median()
)
```

```python
df['Work_Hours_Per_Week'].isnull().sum()
```

```python
# Verify Missing Values Removed

df.isnull().sum()
```

#### DUPLICATE CHECK

```python
duplicates = df.duplicated().sum()

print("Duplicate Records:", duplicates)
```

```python
df.dtypes
```

### FEATURE ENGINEERING

```python
# Create Risk Category

df['Risk_Category'] = pd.cut(
    df['AI_Replacement_Risk'],
    bins=[0,0.3,0.6,1],
    labels=['Low','Medium','High']
)
```

```python
# Showing Risk Category

df[['AI_Replacement_Risk', 'Risk_Category']].head()
```

```python
# Create Salary Category

df['Salary_Category'] = pd.cut(
    df['Average_Salary_USD'],
    bins=3,
    labels=['Low','Medium','High']
)
```

```python
# Showing Salary category

df[['AI_Replacement_Risk', 'Salary_Category']].head()
```

## EXPLORATORY DATA ANALYSIS (EDA)

#### 1. Industry-wise AI Risk

```python
industry_risk = df.groupby('Industry')['AI_Replacement_Risk'].mean().sort_values(ascending=False)

industry_risk
```

#### 2. Job-wise Future Demand

```python
job_demand = df.groupby('Job_Title')['Future_Demand_Score'].mean().sort_values(ascending=False)

job_demand
```

#### 3. Industry-wise Salary

```python
industry_salary = df.groupby('Industry')['Average_Salary_USD'].mean().sort_values(ascending=False)

industry_salary
```

#### 4. AI Tool Usage vs Performance

```python
df.groupby('AI_Tool_Usage')['Performance_Score'].mean()
```

#### 5. Country-wise Salary

```python
country_salary = df.groupby('Country')['Average_Salary_USD'].mean().sort_values(ascending=False)

country_salary
```

## MULTIVARIATE ANALYSIS
